```python
from astropy.io import fits
fx = fits.open('/content/PIU-20080104.fits', memmap=True)
```

```python
fx.info()
```

```python
fx[0].data.shape
```

```python
import pandas as pd

pd.DataFrame(fx[0].data)
```

```python
fx[0].header
```

```python
header = []
for a in fx[0].header.values():
  header.append(str(a))
#  print(a)
```

```python
for i in range(len(header)):
  print(i-8,header[i])
```

```python
source = header[0:8]
header = header[8:-1]
header
```

```python
source
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2,figsize=(16,5))

for name in [x for x in header if 'Amp' in x]:
  ax[0].plot(fx[0].data[:,0],fx[0].data[:,header.index(name)],label=name)
  
for name in [x for x in header if 'Phase' in x]:
  ax[1].plot(fx[0].data[:,0],fx[0].data[:,header.index(name)],label=name)

plt.suptitle(' Station: ' + source[-1].upper() + '\nVLF - DATE OBS ' + source[-2],y=1.1,fontsize=16)
ax[0].set_xlabel(header[0])
ax[1].set_xlabel(header[0])
ax[0].legend(loc='lower left')
ax[1].legend(loc='upper left')
ax[0].grid()
ax[1].grid()
plt.tight_layout()
plt.show()
```
